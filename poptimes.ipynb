```python
import pandas as pd
import numpy as np
import json
```

```python
times = [0] * 24
times
```

```python
with open('btv-bars.json') as f:
    bars = json.loads(f.read())

for bar in bars:
    for day in bar['populartimes']:
        for i in range(0, len(day['data'])):
            times[i] += day['data'][i]

maxtime = max(times)
for i in range(0, len(times)):
    times[i] = times[i] / maxtime * 100

for i, time in enumerate(times):
    print("{0:>2}{1:10.2f}".format(i+1, time))
```
